## Research Question

Is there a statistically significant correlation between ICE arrests and reported rates of rape, murder, robbery, and assault in Texas from 2019 to 2024?

## Background and Prior Work

Immigration and Customs Enforcement (ICE) is a federal agency acting under the Department of Homeland Security. The main responsibilities of ICE include enforcing immigration law, preventing terrorism, and stopping illegal movement of people and goods within the United States. 

The motivation for exploring our topic lies in the dramatic increase in funding that ICE has received with the One Big Beautiful Bill Act. Prior to the passage of the One Big Beautiful Bill Act, ICE received about 10 dollars billion in budget per year. After passing, funding increased to 85 billion dollars. With the marked uptick in funding, activity, and media coverage, ICE has become a hotly contested topic for the American public in recent months, especially with notable incidents involving Renee Good and Alex Pretti. We will attempt to gain a better understanding of the efficacy of ICE on reducing violent and property crime.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

An article published in the The Journal of Law & Economics by Thomas J. Miles and Adam B. Cox examined the relationship between immigration enforcement and local US crime rates. The research specifically investigates the effect of the Secure Communities program. The Secure Communities program uses fingerprint-sharing between local law enforcement and federal immigration authorities to facilitate detention and removal of noncitizens. This is not a direct correlate to ICE, however, it may serve as a proxy to understand whether or not strict enforcement of immigration policy reduces crime. The conclusion of the study showed that the Secure Communities program led to hundreds of thousands of detentions, but did not reduce crime-rates as expected.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

The article Immigration and Customs Enforcement Raids the Pillar of a Community: Student Achievement, Absenteeism, and Mobility Following a Large Worksite Enforcement Operation in North Texas looks at the effects that ICE had through various time extents. It serves as a point of interest as to why Texas was chosen for this project, as well as a way to question what sort of empirical and statistical impact ICE had.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)

Of all states affected by ICE, Texas stands out in the number and scale of people detained. According to data from the Transactional Records Access Clearinghouse (TRAC) Immigration, there are 18,684 people in detention, as of January 22, 2026, which makes Texas the state with the most ICE detainees followed by Louisiana, 8,207 people, and California, 6,422. Geographically, Texas shares the longest border between the U.S. and Mexico, stretching 382 miles, with more than 40 border security stations across 5 sectors. The sheer size of the border and affiliated border control stations are key in understanding the subsequent size of detainees in Texas, which should be considered when examining the correlation between violent crime and ICE arrests. While Texas holds the greatest number of detainees, it is critical to understand that detention location is not the same as arrest location. Texas maintains some of the largest ICE detention centers in the country, where those arrested by ICE both in and out of Texas may be placed into. Therefore it is more representative to examine arrests per capita, to measure the presence of ICE in Texas independent from other states.<a name="cite_ref-4"></a>[<sup>4,</sup>](#cite_note-4)<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5)

## Hypothesis


There will be no statistically significant relationship between the level of ICE arrests in a given month and the rates of rape, murder, robbery, and aggravated assault reported in Texas during that **same month or the following month** (0-month and 1-month lag). We focus on same-month and one-month-lagged comparisons because administrative arrests are recorded at the time of apprehension; if enforcement activity deters crime, an effect would most plausibly appear within a short window. This hypothesis is consistent with prior research on the Secure Communities program, which found no significant crime-reduction effect despite large-scale detentions (Miles & Cox, 2014).

### Dataset #3: Texas Crime Reports, Monthly

```python
# Load the saved clean crime data
texas_offenses_all_monthly = pd.read_csv('data/02-processed/texas_offenses_monthly_clean.csv')
print('\nEach row represents the monthly count for a particular offense type\n')
display(texas_offenses_all_monthly.head())

print("Shape:", texas_offenses_all_monthly.shape)
print("\nData Types:\n", texas_offenses_all_monthly.dtypes)
print("\nNull Counts:\n", texas_offenses_all_monthly.isnull().sum())
print("\nNull Proportions:\n", texas_offenses_all_monthly.isnull().mean().round(4))

# Convert Month column to datetime
texas_offenses_all_monthly['Month'] = pd.to_datetime(texas_offenses_all_monthly['Month'])

# Offenses by year (total)
offenses_by_year = texas_offenses_all_monthly.groupby(
    texas_offenses_all_monthly['Month'].dt.year
)['Count'].sum().astype(int)
offenses_by_year.index.name = 'Year'

print("\nOffenses by Year:\n", offenses_by_year)

# Offense type breakdown (total across all months)
offense_type_total = texas_offenses_all_monthly.groupby('Offense')['Count'].sum().astype(int)
print("\nOffense Type Breakdown:\n", offense_type_total)

# Offense type by year
offense_by_year = texas_offenses_all_monthly.groupby(
    [texas_offenses_all_monthly['Month'].dt.year, 'Offense']
)['Count'].sum().unstack(fill_value=0).astype(int)
offense_by_year.index.name = 'Year'

print("\nOffense Type by Year:")
print(offense_by_year)
```
