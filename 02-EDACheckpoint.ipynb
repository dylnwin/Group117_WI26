# Research Question

Is there a statistically significant correlation between ICE arrests and reported rates of rape, murder, robbery, and assault in Texas from 2019 to 2024?

## Background and Prior Work

Immigration and Customs Enforcement (ICE) is a federal agency acting under the Department of Homeland Security. The main responsibilities of ICE include enforcing immigration law, preventing terrorism, and stopping illegal movement of people and goods within the United States. 

The motivation for exploring our topic lies in the dramatic increase in funding that ICE has received with the One Big Beautiful Bill Act. Prior to the passage of the One Big Beautiful Bill Act, ICE received about 10 billion dollars in budget per year. After passing, funding increased to 85 billion dollars. With the marked uptick in funding, activity, and media coverage, ICE has become a hotly contested topic for the American public in recent months, especially with notable incidents involving Renee Good and Alex Pretti. We will attempt to gain a better understanding of the efficacy of ICE on reducing violent and property crime.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

An article published in the The Journal of Law & Economics by Thomas J. Miles and Adam B. Cox examined the relationship between immigration enforcement and local US crime rates. The research specifically investigates the effect of the Secure Communities program. The Secure Communities program uses fingerprint-sharing between local law enforcement and federal immigration authorities to facilitate detention and removal of noncitizens. This is not a direct correlate to ICE, however, it may serve as a proxy to understand whether or not strict enforcement of immigration policy reduces crime. The conclusion of the study showed that the Secure Communities program led to hundreds of thousands of detentions, but did not reduce crime-rates as expected.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

The article Immigration and Customs Enforcement Raids the Pillar of a Community: Student Achievement, Absenteeism, and Mobility Following a Large Worksite Enforcement Operation in North Texas looks at the effects that ICE had through various time extents. It serves as a point of interest as to why Texas was chosen for this project, as well as a way to question what sort of empirical and statistical impact ICE had.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)

Of all states affected by ICE, Texas stands out in the number and scale of people detained. According to data from the Transactional Records Access Clearinghouse (TRAC) Immigration, there are 18,684 people in detention (as of January 22, 2026), which makes Texas the state with the most ICE detainees, followed by Louisiana, 8,207 people, and California, 6,422. Geographically, Texas shares the longest border between the U.S. and Mexico, stretching 382 miles, with more than 40 border security stations across 5 sectors. The sheer size of the border and affiliated border control stations are key in understanding the subsequent size of detainees in Texas. While Texas holds the greatest number of detainees, it is critical to understand that detention location is not the same as arrest location. Texas maintains some of the largest ICE detention centers in the country, where those arrested both in and out of Texas may be placed into.
```python
import pandas as pd
```

```python
def clean_ice_df(ice_df):
    # fix column names
    ice_df.columns = ice_df.iloc[3]

    # remove unecessary rows from beginning
    ice_df_drop = ice_df.iloc[4:].reset_index(drop=True)

    # keep only date and AOR columns
    ice_df_drop = ice_df_drop[['Apprehension AOR', 'Apprehension Date And Time']]

    # keep only Texas arrests
    texas_aor = ['Dallas Area of Responsibility', 
             'El Paso Area of Responsibility', 
             'Harlingen Area of Responsibility', 
             'Houston Area of Responsibility', 
             'San Antonio Area of Responsibility']
    ice_df_texas = ice_df_drop[ice_df_drop['Apprehension AOR'].isin(texas_aor)]

    #filter for years 2019-2024
    ice_df_texas['Apprehension Date And Time'] = pd.to_datetime(
    ice_df_texas['Apprehension Date And Time'], errors='coerce')
    ice_df_2019_2024 = ice_df_texas[
    ice_df_texas['Apprehension Date And Time'].dt.year.between(2019, 2024)]

    #remove any nan rows
    ice_df_clean = ice_df_2019_2024.dropna()
    
    return ice_df_clean
```

```python
# years of each data file
years = [[2018, 2019],
 [2019, 2020],
 [2020, 2021],
 [2021, 2022],
 [2022, 2023],
 [2023, 2024],
 [2024, 2025]]

clean_dfs = []  # list to store cleaned dataframes

for yr in years:
    year_start, year_end = yr

    # load data and clean it 
    ice_df = pd.read_excel(f'data/00-raw/ice_arrests/ice_{year_start}-{year_end}.xlsx')
    clean_ice = clean_ice_df(ice_df)

    # append to the list
    clean_dfs.append(clean_ice)

# concatenate all cleaned dataframes into one
final_clean_ice = pd.concat(clean_dfs, ignore_index=True)
```

## Texas Crime Reports - Monthly

This dataset is from the same source as the previous dataset. Our group has decided to add this dataset to get finer-grain detail on the rise and fall of crime rates on a monthly basis, rather than yearly. The variables of this dataset are the type of crime, the month the crime was committed in, and number of attempted, completed, and total crimes for the month. This data will be used to track crime rates per month, and can be broken down further to analyze the type of crime committed per month. 

```python
import pandas as pd
import os

# Load dataframe from 00-raw
monthly_crime_df = pd.read_csv('data/00-raw/texas_crime_reports_monthly/texas_offenses_monthly.csv')

# Forward-fill offense names
monthly_crime_df['Offense'] = monthly_crime_df['Offense'].ffill()

# Keep only required offense codes
target_codes = ['11A', '09A', '120', '13A']
monthly_crime_df = monthly_crime_df[
    monthly_crime_df['Offense'].str.contains('|'.join(target_codes), na=False)
].copy()

# Map to clean offense names
offense_labels = {
    '11A': 'Rape',
    '09A': 'Murder and Nonnegligent Manslaughter',
    '120': 'Robbery',
    '13A': 'Aggravated Assault'
}

monthly_crime_df['OffenseCode'] = monthly_crime_df['Offense'].str.extract(r'(11A|09A|120|13A)')
monthly_crime_df['Offense'] = monthly_crime_df['OffenseCode'].map(offense_labels)

# Clean month column
monthly_crime_df['Month'] = pd.to_datetime(
    monthly_crime_df['Month'],
    format='%b - %Y'
)

# Remove commas from numeric columns
cols = ['Attempted', 'Completed', 'Total']
for col in cols:
    monthly_crime_df[col] = (
        monthly_crime_df[col]
        .astype(str)
        .str.replace(',', '', regex=False)
    )

# Convert numeric columns
monthly_crime_df[cols] = monthly_crime_df[cols].apply(
    pd.to_numeric,
    errors='coerce'
)

# Convert to tidy format
tidy_crime_df = monthly_crime_df.melt(
    id_vars=['Month', 'Offense'],
    value_vars=['Attempted', 'Completed'],
    var_name='Outcome',
    value_name='Count'
).dropna(subset=['Count'])

# Sort
tidy_crime_df = tidy_crime_df.sort_values(['Offense', 'Month'])

# Save cleaned file
output_dir = "data/02-processed"
os.makedirs(output_dir, exist_ok=True)

output_path = os.path.join(output_dir, "texas_offenses_monthly_clean.csv")
tidy_crime_df.to_csv(output_path, index=False)
```

```python
def plot_time_series_monthly(monthly_crime_df, crime_type):
    # function for plotting distribution of crime over time for the monthly data

    # convert to datetime
    monthly_crime_df['Month'] = pd.to_datetime(monthly_crime_df['Month'])

    # combine all crime categories per month
    monthly_totals = monthly_crime_df.groupby('Month')['Count'].sum().reset_index()

    # plot
    plt.figure()
    plt.plot(monthly_totals['Month'], monthly_totals['Count'])
    plt.xlabel('Date')
    if crime_type == 'Crime (all)':
        plt.ylabel('Number of Crimes (all)')
        plt.title(f'{crime_type} Per Month')
    elif crime_type == 'Robbery':
        plt.ylabel('Number of Robberies')
        plt.title('Robberies Per Month')
    else:
        plt.ylabel(f'Number of {crime_type}s')
        plt.title(f'{crime_type}s Per Month')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# load data
monthly_crime_df = pd.read_csv('data/02-processed/texas_offenses_monthly_clean.csv')

# plot time series of crime overall
plot_time_series_monthly(monthly_crime_df, 'Crime (all)')

# plot time series for each type of crime
crime_categories = ['Rape', 'Murder and Nonnegligent Manslaughter', 'Robbery', 'Aggravated Assault']

for crime_type in crime_categories:
    crime_df_new = monthly_crime_df[crime_df['offense_code'] == crime_type]
    plot_time_series_monthly(crime_df_new, crime_type)
```

```python
def plot_correlation_monthly(ice_df, monthly_crime_df, crime_type):
    # function for plotting correlation between ICE arrests and crime

    # find counts of ICE arrests and crime by year
    ice_df['datetime'] = pd.to_datetime(ice_df['Apprehension Date And Time'])
    ice_monthly = ice_df.set_index('datetime').resample('M').size()
    ice_monthly.index = ice_monthly.index.year


    # convert to datetime
    monthly_crime_df['Month'] = pd.to_datetime(monthly_crime_df['Month'])

    # combine all crime categories per month
    crime_monthly = monthly_crime_df.groupby('Month')['Count'].sum().reset_index()


    ice_counts = ice_monthly.reset_index(drop=True)
    crime_monthly = crime_monthly.sort_values('Month').reset_index(drop=True)
    crime_monthly['ice_arrests'] = ice_counts
    
    # plotting
    sns.regplot(x='ice_arrests', y='Count', data=crime_monthly)

    plt.title(f'Correlation Between ICE Arrests and {crime_type}')
    plt.xlabel('ICE Arrests')
    plt.ylabel(f'{crime_type} Count')
    plt.show()

    # print out correlation coefficient
    corr_value = crime_monthly['ice_arrests'].corr(crime_monthly['Count'])
    print(f'Correlation coefficient is {corr_value:.3f}')
```

```python
# load data
monthly_crime_df = pd.read_csv('data/02-processed/texas_offenses_monthly_clean.csv')
ice_df = pd.read_csv('data/02-processed/ice_arrests_clean.csv')

# plot correlation between ICE arrests and crime overall
plot_correlation_monthly(ice_df, monthly_crime_df, 'All Crime')

# plot correlation between ICE arrests and each type of crime
crime_categories = ['Rape', 'Murder and Nonnegligent Manslaughter', 'Robbery', 'Aggravated Assault']

for crime_type in crime_categories:
    crime_df_new = monthly_crime_df[crime_df['offense_code'] == crime_type]
    plot_correlation_monthly(ice_df, crime_df_new, crime_type)
```
